```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
ruta = '/content/drive/MyDrive/Colab_Notebooks/Airdata.csv'
df = pd.read_csv(ruta)
df.tail()
```

```python
df.info()

print(df.dtypes)

print(df.isnull().sum())
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%m/%d/%Y %H:%M')

df['Año'] = df['DateTime'].dt.year
df['Mes'] = df['DateTime'].dt.month
df['Dia'] = df['DateTime'].dt.day
df['Hora'] = df['DateTime'].dt.hour
df['Minuto'] = df['DateTime'].dt.minute

df.head()
```

```python
df['DateTime_Modificado'] = df['DateTime'] + pd.Timedelta(days=10, hours=7, minutes=15)

print("Fecha original vs Fecha modificada (+10 días, 7 horas, 15 minutos):")
print(df[['DateTime', 'DateTime_Modificado']].head(10))

df['Dia_Semana_Original'] = df['DateTime'].dt.day_name()
df['Dia_Semana_Modificado'] = df['DateTime_Modificado'].dt.day_name()

print("\nCon detalles de días de la semana:")
df[['DateTime', 'Dia_Semana_Original', 'DateTime_Modificado', 'Dia_Semana_Modificado']].head(10)
```

```python
df.head()
```

```python
df['Unix_Timestamp'] = df['DateTime'].astype('int64')

print("Primeras 3 filas - DateTime vs Unix Timestamp:")
print(df[['DateTime', 'Unix_Timestamp']].head(3))
```

```python
anio_filtro = 2020
mes_filtro = 11
dia_filtro = 1

df_filtrado = df[
    (df['Año'] > anio_filtro) |
    ((df['Año'] == anio_filtro) & (df['Mes'] > mes_filtro)) |
    ((df['Año'] == anio_filtro) & (df['Mes'] == mes_filtro) & (df['Dia'] > dia_filtro))
]

print(f"Filas después del filtro: {len(df_filtrado)}")
df_filtrado.head()
```
