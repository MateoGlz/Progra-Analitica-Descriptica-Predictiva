```python
import pandas as pd
import numpy as np
ruta = '/content/drive/MyDrive/Colab_Notebooks/nls97.csv'
df = pd.read_csv(ruta)
df.head()
```

```python
print("Valores únicos:", df['maritalstatus'].unique())

print("\nFrecuencia:")
print(df['maritalstatus'].value_counts())
```

```python
df['maritalstatus'] = df['maritalstatus'].replace('Never-married', 'Single')

print("Valores únicos después del cambio:")
print(df['maritalstatus'].unique())

print("\nFrecuencia después del cambio:")
print(df['maritalstatus'].value_counts())
```

```python
import numpy as np

df['marital_grouped'] = np.where(
    df['maritalstatus'].isin(['Single', 'Widowed']),
    'Single/No Partner',
    df['maritalstatus']
)

print("Valores únicos de marital_grouped:")
print(df['marital_grouped'].unique())

print("\nFrecuencia:")
print(df['marital_grouped'].value_counts())
```

```python

print("Tipo de datos de la columna 'gender':")
print(df['gender'].dtype)

print("\nCantidad de bytes de la columna 'gender':")
print(df['gender'].memory_usage(deep=True))
```

```python
df['gender'] = df['gender'].astype('category')

print("Tipo de datos después de la conversión:")
print(df['gender'].dtype)

print("\nCantidad de bytes después de la conversión:")
print(df['gender'].memory_usage(deep=True))

print("\nCategorías únicas:")
print(df['gender'].cat.categories)
```
