```python
# 1. Cargar datos y verificar
df = pd.read_csv(ruta)
print("=== EJERCICIO 05: CÁLCULO DE MEDIAS CON Y SIN DUPLICADOS ===")
print("\nPrimeras filas del dataset:")
print(df.head())
print(f"\nCantidad total de registros originales: {len(df)}")

# 2. Calcular e imprimir la media de Pulse y Maxpulse (CON duplicados)
print("\n--- MEDIAS CON DUPLICADOS ---")
media_pulse_con = df['Pulse'].mean()
media_maxpulse_con = df['Maxpulse'].mean()

print(f"Media de Pulse (con duplicados): {media_pulse_con:.2f}")
print(f"Media de Maxpulse (con duplicados): {media_maxpulse_con:.2f}")

# 3. Usar el conjunto SIN duplicados que ya creamos (df_keep_last del Ejercicio 03)
# Recordemos que en el Ejercicio 03 creamos:
df_keep_last = df.drop_duplicates(subset=['Pulse', 'Maxpulse'], keep='last')

print(f"\nRegistros originales: {len(df)}")
print(f"Registros sin duplicados (keep='last'): {len(df_keep_last)}")
print(f"Duplicados eliminados: {len(df) - len(df_keep_last)}")

# 4. Calcular medias SIN duplicados (usando df_keep_last)
print("\n--- MEDIAS SIN DUPLICADOS ---")
media_pulse_sin = df_keep_last['Pulse'].mean()
media_maxpulse_sin = df_keep_last['Maxpulse'].mean()

print(f"Media de Pulse (sin duplicados): {media_pulse_sin:.2f}")
print(f"Media de Maxpulse (sin duplicados): {media_maxpulse_sin:.2f}")

# 5. Comparar resultados
print("\n--- COMPARACIÓN ---")
diferencia_pulse = media_pulse_sin - media_pulse_con
diferencia_maxpulse = media_maxpulse_sin - media_maxpulse_con

print(f"Diferencia en Pulse: {diferencia_pulse:+.2f}")
print(f"Diferencia en Maxpulse: {diferencia_maxpulse:+.2f}")

print(f"\nVariación porcentual en Pulse: {(diferencia_pulse/media_pulse_con)*100:+.2f}%")
print(f"Variación porcentual en Maxpulse: {(diferencia_maxpulse/media_maxpulse_con)*100:+.2f}%")

# 6. Mostrar por qué cambió (análisis de los duplicados eliminados)
print("\n--- ANÁLISIS DE LOS DUPLICADOS ELIMINADOS ---")
# Identificar qué valores se eliminaron
duplicados = df[df.duplicated(subset=['Pulse', 'Maxpulse'], keep='last')]
print("Registros duplicados que fueron eliminados (se mantuvo el último):")
print(duplicados[['Pulse', 'Maxpulse']])

if len(duplicados) > 0:
    print(f"\nMedia de Pulse en duplicados eliminados: {duplicados['Pulse'].mean():.2f}")
    print(f"Media de Maxpulse en duplicados eliminados: {duplicados['Maxpulse'].mean():.2f}")
```
