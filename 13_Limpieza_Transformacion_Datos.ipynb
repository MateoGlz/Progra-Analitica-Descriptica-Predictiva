```python
import pandas as pd
import numpy as np
ruta = '/content/drive/MyDrive/Colab_Notebooks/miaad-nyc-r-s.csv'
# df = pd.read_csv(ruta)
df = pd.read_csv(ruta, skipfooter=1, engine='python')
df.head()
```

```python
df.tail(3)
```

```python
print(f"El DataFrame tiene {df.shape[0]} filas y {df.shape[1]} columnas")
```

```python
nombre_actual = df.columns[0]

df = df.rename(columns={nombre_actual: 'INDEX MIIAD'})

print(f"Nombre cambiado: '{nombre_actual}' → 'INDEX MIIAD'")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
```

```python
nombres_originales = df.columns.tolist()

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^\w_]', '', regex=True)

print("✅ Nombres de columnas normalizados:")
print(f"De: {nombres_originales[:5]}...")
print(f"A:  {df.columns.tolist()[:5]}...")
```

```python
columnas_a_verificar = ['sale_price', 'land_square_feet', 'gross_square_feet']

def convertir_a_float(serie):
    return pd.to_numeric(serie.astype(str).str.replace(',', '').str.replace('$', ''),
                         errors='coerce')

print("\nCONVIRTIENDO A FLOAT...")
for columna in columnas_a_verificar:
    if columna in df.columns:
        nulos_antes = df[columna].isnull().sum()

        df[columna] = convertir_a_float(df[columna])

        nulos_despues = df[columna].isnull().sum()
        nuevos_nulos = nulos_despues - nulos_antes

        print(f"{columna}: convertido a {df[columna].dtype}")
        if nuevos_nulos > 0:
            print(f"Se generaron {nuevos_nulos} nuevos valores nulos")
    else:
        print(f"{columna}: Columna no encontrada")
```

```python
print("\n=== VALORES NULOS ===")
print(df.isnull().sum())
print("\n" + "="*50)

print("=== PRIMERAS 5 FILAS ===")
print(df.head())
print("\n" + "="*50)

print("\n=== INFORMACIÓN DEL DATASET ===")
print(df.info())
print("\n" + "="*50)

print("\n=== ESTADÍSTICAS DESCRIPTIVAS ===")
print(df.describe())
print("\n" + "="*50)

print("\n=== DISTRIBUCIÓN DE VARIABLES CATEGÓRICAS ===")
print("\nBUILDING CLASS CATEGORY:")
print(df['building_class_category'].value_counts())
print("\nNEIGHBORHOOD:")
print(df['neighborhood'].value_counts())
print("TAX CLASS AT PRESENT:")
print(df['tax_class_at_present'].value_counts())
print("\nBUILDING CLASS AT PRESENT:")
print(df['building_class_at_present'].value_counts())
print("\nADDRESS:")
print(df['address'].value_counts())
```
