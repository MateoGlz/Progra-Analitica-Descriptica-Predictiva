```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    x=df2007['gdpPercap'],
    y=df2007['lifeExp'],
    s=df2007['pop']/2250000,
    c=df2007['continent'].astype('category').cat.codes,
    cmap='Set1',
    alpha=0.7
)

plt.xlabel('PIB per cápita')
plt.ylabel('Esperanza de vida')
plt.title('PIB vs Esperanza de vida por Continente')

continentes = df2007['continent'].unique()
handles, labels = scatter.legend_elements(prop="colors")
plt.legend(handles, continentes, title="Continentes")

plt.grid(True, alpha=0.3)
plt.show()
```
